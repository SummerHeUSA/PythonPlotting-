```python
city = pd.read_csv('/Users/summerhe/Desktop/UDEN201811DATA3/Week5_PythonPlotting/HW/Pyber/data/city_data.csv')
ride = pd.read_csv('/Users/summerhe/Desktop/UDEN201811DATA3/Week5_PythonPlotting/HW/Pyber/data/ride_data.csv')
```

```python
pd.merge(ride,city, on="city")
```

```python
combined_data= pd.merge(city, ride, how='outer', on='city')

combined_data.head()
```

```python
combined_data.head(2)
```

```python
color_dict={"Urban":"red","Suburban":"blue","Rural":"green"}
out = combined_data.groupby(['city',"type"]).agg({'ride_id':'count','fare':'mean'}).reset_index()
out["Color"] = out["type"].apply(lambda x: color_dict[x])
out.head()
```

```python
out.head()
```

```python
# out.plot(kind = 'scatter', x ='ride_id', y ='fare', s = combined_data.driver_count*10, alpha=.33, c =out["Color"],label=["A","B","C"])

plt.scatter(out[out["type"]=="Urban"]["ride_id"],out[out["type"]=="Urban"]["fare"],s = combined_data.driver_count*10,c="r",alpha=.33,label="Urban")
plt.scatter(out[out["type"]=="Suburban"]["ride_id"],out[out["type"]=="Suburban"]["fare"],s = combined_data.driver_count*10,c="b",alpha=.33,label="Suburban")
plt.scatter(out[out["type"]=="Rural"]["ride_id"],out[out["type"]=="Rural"]["fare"],s = combined_data.driver_count*10,c="g",alpha=.33,label="Rural")
plt.legend()
plt.show()
```
